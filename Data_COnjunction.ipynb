```python
import pandas as pd

# --- Part 1: Check the metadata for the sample file ---
sample_metadata_path = '/content/drive/MyDrive/brca_metabric/meta_clinical_sample.txt'

print(f"--- Contents of {sample_metadata_path} ---")
try:
    with open(sample_metadata_path, 'r') as f:
        print(f.read())
except FileNotFoundError:
    print(f"File not found: {sample_metadata_path}")


# --- Part 2: Load the data and inspect its columns ---
sample_data_path = '/content/drive/MyDrive/brca_metabric/data_clinical_sample.txt'

print(f"\n--- Column summary for {sample_data_path} ---")
try:
    # Load the data, assuming it's tab-separated and ignoring comment lines
    sample_df = pd.read_csv(sample_data_path, sep='\t', comment='#')
    sample_df.info()
except FileNotFoundError:
    print(f"File not found: {sample_data_path}")
```

```python
# Assuming 'clinical_df' is your loaded patient data
# and 'sample_df' is your loaded sample data

# Merge the two dataframes on the 'PATIENT_ID' column
combined_clinical_df = pd.merge(clinical_df, sample_df, on='PATIENT_ID', how='left')

# Inspect the newly merged dataframe
print("--- Summary of the Combined Clinical DataFrame ---")
combined_clinical_df.info()

print("\n--- First 5 rows of the Combined DataFrame ---")
print(combined_clinical_df.head())
```

```python
# Assuming 'combined_clinical_df' is your merged clinical data from the last step

# 1. Load the mRNA expression data
mrna_file_path = '/content/drive/MyDrive/brca_metabric/data_mrna_illumina_microarray.txt'
mrna_df = pd.read_csv(mrna_file_path, sep='\t')

# 2. Set the gene names as the index. We assume the column is 'Hugo_Symbol'
mrna_df = mrna_df.set_index('Hugo_Symbol')

# 3. Transpose the DataFrame so samples are rows and genes are columns
mrna_df_transposed = mrna_df.transpose()

# 4. Merge with our main clinical dataframe using SAMPLE_ID
# We use the clinical df's 'SAMPLE_ID' column and the transposed mRNA df's index
final_df = pd.merge(combined_clinical_df, mrna_df_transposed,
                    left_on='SAMPLE_ID', right_index=True, how='left')

# Inspect the final, massive DataFrame
print("--- Shape of the Final Combined DataFrame ---")
print(final_df.shape)

print("\n--- A peek at the final DataFrame ---")
print(final_df.head())
```
